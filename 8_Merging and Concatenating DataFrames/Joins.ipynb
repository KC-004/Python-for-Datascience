## 1.3 Exclusive Outer Join (Exclusive Full join)

```python
df1 = pd.DataFrame({'id': ['A', 'B', 'C', 'D'],
                    'age': [30, 23, 25, 22]})
    
df2 = pd.DataFrame({'id': ['C', 'D', 'E','F'],
                    'job': ['Doctor', 'Statistician', 'Accountant',
                            'Developer']})
```

```python
df1
```

```python
df2
```

```python
# indicator=True
df1.merge(df2, on='id', how='outer', indicator=True)
```

```python
# query

df_only_left_right = df1.merge(df2, on='id', how='outer', indicator=True).query("_merge=='left_only' or _merge=='right_only'")
df_only_left_right 
```

```python
# shape

df_only_left_right.shape
```
