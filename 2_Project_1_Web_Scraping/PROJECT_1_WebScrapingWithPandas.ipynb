# 2 Read .csv from Multiple URLs with Pandas

Link: root +season + league

```python
# link structure

"https://www.football-data.co.uk/mmz4281/" + "2122" + "/" + "E0.csv"
```

```python
# create a root variable

root = "https://www.football-data.co.uk/mmz4281/"
```

# 2.1 Multiple leagues
https://www.football-data.co.uk/mmz4281/2122/E0.csv
https://www.football-data.co.uk/mmz4281/2021/E1.csv
https://www.football-data.co.uk/mmz4281/2021/E2.csv
https://www.football-data.co.uk/mmz4281/2021/E3.csv
https://www.football-data.co.uk/mmz4281/2021/EC.csv

```python
# create list of leagues

leagues = ['E0', 'E1', 'E2', 'E3', 'EC']
frames = []
#looping through leagues, read multiple csv and append it to it

for league in leagues:
    df =  pd.read_csv(root + '2122' + '/' + league + '.csv', encoding = 'unicode_escape')
    frames.append(df)
```

```python
# length of frames
len(frames)
```

```python
# show 1st, 2nd and 3rd element

frames[3]
```

## 2.2 Multiple Seasons

```python
for season in range(15,21):
    print(str(season) + str(season + 1))
```

```python
# creating list of leagues
leagues = ['E0', 'E2', 'E3']
frames = []

#loop through range of seasons and insert new column

for league in leagues:
    for season in range(15, 21):
        df =  pd.read_csv(root + '2122' + '/' + league + '.csv')
        df.insert(1, 'season', season)
        frames.append(df)

    
```

# 3 Organize All the Data in a Dictionary

```python
# create a dictionary with original name of league as key

dict_countries = {
            'Spanish La Liga':'SP1', 'Spanish Segunda Division':'SP2',
            'German Bundesliga':'D1',
            'English Premier league':'E0',
            'English league 1':'E2',
            'English league 2':'E3'
}
```

```python
dict_countries.keys()
```

```python
for league in dict_countries:
    print(dict_countries[league])
```

```python

dict_historical_data = {}


for league in dict_countries:
    frames = []
    for season in range(15, 21):
        df =  pd.read_csv(root + '2122' + '/' + dict_countries[league] + '.csv')
        df.insert(1, 'season', season)
        frames.append(df)
    df_concat = pd.concat(frames)
    dict_historical_data[league] = df_concat
    
```
