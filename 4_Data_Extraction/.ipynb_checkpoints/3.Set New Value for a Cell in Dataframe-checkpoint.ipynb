```python
import pandas as pd
```

```python
df = pd.read_csv('players_20.csv')
df.set_index('short_name', inplace=True)
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality']]
```

```python
df.head()
```

# 1 Set Value to on cell


```python
# update L.Messi height

df.loc['L. Messi', 'height_cm'] = 175
```

```python
df
```

# 2 set value to entire column

```python
# set height of all players to 190

df.loc[:, 'height_cm'] = 190
```

```python
df
```

# 3 Set value to entire row

```python
# get all columns that correspond to player ranked last in fifa

df.iloc[[-1], :]
```

```python
import numpy as np

np.nan
```

```python
df.iloc[-1, :] = np.nan
```

```python
df
```

 # 4 Set value to multiple cells

```python
#set value for all items matchning the list of labels

df.loc[['L. Messi', 'Cristiano Ronaldo'], ['height_cm']] = 175
```

```python
df
```

# 5 Set value for row matching a condition

```python
# set value for rows matching a condition

columns = ['age', 'dob', 'height_cm', 'weight_kg']

df.loc[(df['height_cm']>180), columns] = 0
```

```python
df
```
