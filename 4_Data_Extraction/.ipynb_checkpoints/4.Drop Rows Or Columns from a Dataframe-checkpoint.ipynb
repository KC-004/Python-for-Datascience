```python
import pandas as pd
```

```python
df = pd.read_csv('players_20.csv')
df.set_index('short_name', inplace=True)
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality']]
```

```python
df.head()
```

# 1 Drop Row

```python
# drop on row
# axis parameter

df.drop('L. Messi', axis=0) # 0 represents row and 1 for column
```

```python
# index parameter

df.drop(index=['L. Messi'])
```

```python
# drop two or more rows and update data (inplace = True)

df.drop(['L. Messi', 'Neymar Jr'], axis=0, inplace=True)
```

```python
df
```

# 2 Drop Columns

```python
# drop one column
# axis parameter

df.drop('long_name', axis=1 )
```

```python
# index parameter

df.drop(columns=['long_name'])
```

```python
# drop column by position/index (drop last column)

df.drop(df.columns[[-1]], axis=1)
```

```python
# drop tow ro more columns and update df

df.drop(['long_name','dob'], axis=1, inplace=True)
```
