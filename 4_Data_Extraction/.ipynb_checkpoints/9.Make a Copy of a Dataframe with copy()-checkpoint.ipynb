```python
import pandas as pd
```

```python
df = pd.read_csv('players_20.csv')
df.set_index('short_name', inplace=True)
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
df.head()
```

# 1 copy()


## 1.1 deep = True

```python
# deep = True by default (Modifications to the data or indices of the copy will not change)

df_copy = df.copy()
```

```python
# update value in original dataframe

df.loc['L. Messi', 'height_cm'] = 180
```

```python
df_copy
```

## 1.2 deep = False

```python
# deep=False (Any changes to the data of the original will be reflected in the copy)

df_copy = df.copy(deep=False)
```

```python
# update value in original dataframe

df.loc['Cristiano Ronaldo', 'height_cm'] = 200
```

```python
df_copy
```

## 1.3 Simple assignment

```python
# make a copy

df_new_copy = df 
```

```python
# update value in original dataframe

df.loc['Neymar Jr', 'height_cm'] = 190
```

```python
# copy vs original dataframe

df_new_copy
```
