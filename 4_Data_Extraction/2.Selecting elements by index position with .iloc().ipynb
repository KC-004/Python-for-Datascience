```python
import pandas as pd
```

```python
df = pd.read_csv('players_20.csv')
df.set_index('short_name', inplace=True)
```

```python
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
df.head()
```

# 1 Selecting with a single value
iloc[row_positin, column_position]

```python
# get the height f L.Messi

df.iloc[0, 3]
```

```python
# get the weight of cristiano ronaldo

df.iloc[1, 4]
```

```python
# get all the rows inside the 'height_cm column'

df.iloc[:, [0, 3]]
```

```python
# get all columns that corresponds to index 'L. Messi'

df.iloc[[0], :]
```

# 2 Selecting with a list of values


```python
# get all data about L.Messi and cristiano ronaldo

df.iloc[[0, 1], :]
```

```python
# get the height of Messi and Ronaldo

df.iloc[[0, 1], 3]
```

```python
# get the height and weight of messi

df.iloc[0, [3, 4]]
```

```python
# get height and weight of messi and ronaldo

df.iloc[[0, 1], [3, 4]]
```

# 3 Selecting a range of data with a slice
start:stop (Note that the start is included but the stop is excluded)

```python
# slice column labels age/club

players = [0, 1]

df.iloc[players, 1:6+1]
```

```python
# slice index labels (top1 to top10 player)
#columns: age, dob, height_cm, weight_kg

df.iloc[:10, 1:5]
```

# 4 Selecting with conditions
iloc cannot accept a boolean series, but only a boolean list. We have to use the list() function to convert a series into a boolean list.

```python
# one condition: select player with height above 180cm

columns = [1,2,3,4,5]

list(df['height_cm']>180)

df.iloc[list(df['height_cm']>180), columns]
```

```python
# multiple conditions: select player with height above 180cm and from Argentina

condition = list((df['height_cm']>180) & (df['nationality']=='Argentina'))

df.iloc[condition, columns]
```
