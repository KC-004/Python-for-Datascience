```python
import pandas as pd
```

```python
df = pd.read_csv('players_20.csv')
df.set_index('short_name', inplace=True)
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
# extract 10 ranom elements from the "nationality" column 

import random
import numpy as np

df['nationality'].sample(10, random_state=99)
```

```python
# extranct a random 20%  sample of the dataframe

df.sample(frac=0.2, random_state=12)
```

```python
# upsample: increase the sampling rate -> frac>1 

df.sample(frac=2, replace=True, random_state=34) #doubles the number of rows
```
