```python
import pandas as pd
```

```python
df_laptops = pd.read_csv('laptop_price.csv', encoding='latin-1')
df_laptops.head(3)
```

# 1 drop_duplicates()

```python
# drop duplicates in 2 or more columns

df_laptops.drop_duplicates(['Company'])[['Company', 'Price_euros']]
```

```python
#sort dataframe asending by company and price
# cheapest first and most expensive last

df_laptops = df_laptops.sort_values(['Company', 'Price_euros'])
```

```python
# cheapest : keep='first'

df_laptops.drop_duplicates(['Company'], keep='first')[['Company', 'Price_euros']]
```

```python
# most expensive: keep='last'

df_laptops.drop_duplicates(['Company'], keep='last', inplace=True, ignore_index=True)
df_laptops[['Company', 'Price_euros']]
```

Task: Find the biggest and smallest screen size in laptops of each company using the sort_values() and duplicates() metods (keep "first", "last", False)

```python
# sort dataframe asending by company and inches
# smallet first and biggest last

df_laptops = pd.read_csv('laptop_price.csv', encoding='latin-1')

df_laptops.sort_values(['Company', 'Inches'], inplace=True)
df_laptops[['Company', 'Inches', 'Price_euros']]
```

```python
# smallest screen in each company
df_laptops.drop_duplicates(['Company'], keep='first')[['Company', 'Inches', 'Price_euros']]
```

```python
# laptop with largest screen in each company

df_laptops.drop_duplicates(['Company'], keep='last')[['Company', 'Inches', 'Price_euros']]
```
